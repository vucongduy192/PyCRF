```python
import numpy as np
import pickle

from collections import defaultdict
```

```python
data_file = open('letter.data', 'r')
dictionary = defaultdict() # key : character, value : set of pixel array variants for that character

for line in data_file:
    data_arr = line.split()
    # ---- Label -----
    char = data_arr[1]
    char_label = ord(char) - 97
    # print (char, char_label)

    # ---- Features -----
    char_features = tuple(map(int, data_arr[6:]))
    # print (char_features)
    char_features_set = dictionary.get(char, set())
    char_features_set.add(char_features)
    
    dictionary[char] = char_features_set
```

```python
for key, val in dictionary.items():
    print ("Character : " + str(key))
    print ("Number of variants: " + str(len(val)))


with open("./letter_variants.npy", "wb") as f:
    pickle.dump(dictionary, f)
```

```python
import re
import nltk
```

```python
def get_token_gt_length(book_name, min_len):
    book_tokens = nltk.corpus.gutenberg.words(book_name)
    # remove space or tab ... in head (or tail) of string
    book_tokens = map(lambda x : x.strip().lower(), book_tokens)
    # remove special character of string
    book_tokens = filter(lambda x : re.match(r'^[a-zA-Z]+$', x), book_tokens)
    book_tokens = filter(lambda x : len(x) >= min_len, book_tokens)
    return list(book_tokens)
    
book_name = 'melville-moby_dick.txt'
min_len = 3 # token min length
mb_dick_tokens = get_token_gt_length('melville-moby_dick.txt', min_len)

print (mb_dick_tokens[:10])
print ("Number of extracted tokens " + str(len(mb_dick_tokens)))
```

```python
from collections import Counter

def filter_words_for_stratification(words_list, min_times):
    print (len(words_list))
    print (len(np.unique(words_list)))

    len_words_list = map(lambda x : len(x), words_list)
    len_words_counter = Counter(len_words_list)
    print (len_words_counter)
    not_allowed_lengths = [x[0] for x in len_words_counter.items() if x[1] < 3]
    print (not_allowed_lengths) # tokens has appear times < 3

    allowed_words_list = list(filter(lambda x : len(x) not in not_allowed_lengths, words_list))
    print (len(allowed_words_list))
    print (len(np.unique(allowed_words_list)))
    
    return allowed_words_list
```
